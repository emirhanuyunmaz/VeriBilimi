```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,80,86,85,87,70,90,95]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yaa Gre Deiim Grafii")
plt.show() # Notebook dnda kullanlmas gerek.
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.subplot(1,2,2) # Bir sra olacak iki kolon olacak u an ikicinin verileri girilecek
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
```

```python
benimFigure = plt.figure()
benimAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
benimAxes.plot(numpyDizisi1,numpyDizisi2,"g")
benimAxes.set_xlabel("X Ekseni Veri smi")
benimAxes.set_ylabel("Y Ekseni Veri smi")
benimAxes.set_title("Grafik bal")
```

```python
benimFigure = plt.figure()
benimAxes = benimFigure.add_axes([0.2,0.2,0.4,0.4])
benimAxes.plot(numpyDizisi1,numpyDizisi2,"g")
benimAxes.set_xlabel("X Ekseni Veri smi")
benimAxes.set_ylabel("Y Ekseni Veri smi")
benimAxes.set_title("Grafik bal")
```
